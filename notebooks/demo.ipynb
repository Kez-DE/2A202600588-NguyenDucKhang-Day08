Notebook demo cho buổi trình bày. Hiện notebook kiểm tra nhanh dữ liệu Task 1-3 đã được chuẩn bị trong repo.

```python
from pathlib import Path

ROOT = Path('..').resolve()
legal_files = sorted((ROOT / 'data' / 'landing' / 'legal').glob('*.pdf'))
news_files = sorted(p for p in (ROOT / 'data' / 'landing' / 'news').glob('*.json') if p.name != 'news_sources.json')
markdown_files = sorted((ROOT / 'data' / 'standardized').rglob('*.md'))

print(f'Legal raw files: {len(legal_files)}')
print(f'News crawled files: {len(news_files)}')
print(f'Standardized markdown files: {len(markdown_files)}')
```
